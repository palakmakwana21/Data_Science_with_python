```python
#file automatically closes after the block of code is executed
with open("data.txt","r") as file:
    content = file.read()
    print(content)
```

```python
#read specific number of characters
file = open("data.txt","r")
print(file.read(10))#read first 10 characters
file.close()
```

```python
#readline one line at a time
file = open("data.txt","r")
print(file.readline())#read first line
file.close()
```

```python
file = open("data.txt","r")
print(file.name) #name of the file 
print(file.mode) #mode in which the file is opened
print(file.closed) #check if the file id closed or not
file.read(5)
print(file.tell()) #current position of the file cursor 
file.seek(0) #move the file cursor to the beginning of the file 
```
