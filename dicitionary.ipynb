```python
#nestad dicitionary
company = {
     'emp001':{
    "name" : "rahul",
    "dept" :"hr",
    "salary":4000
    },
    'emp002':{
    "name":"palak",
    "dept":"it",
    "salary":6000
    }
}
print(company)
```
