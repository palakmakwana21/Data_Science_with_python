```python
from random import random, randint, shuffle, choice
```

```python
from datetime import datetime as dt
now = dt.now()
print(now)
print("Current date and time: ", dt.now())
print("current date is: ", dt.now().date())
print("current time is: ", dt.now().time())
print("current year is: ", dt.now().year)
print("current month is: ", dt.now().month)
print("current day is: ", dt.now().day)
print("current hour is: ", dt.now().hour)
print("current minute is: ", dt.now().minute)
print("current second is: ", dt.now().second)
print("current microsecond is: ", dt.now().microsecond)
print("current weekday is: ", dt.now().weekday())
print("current iso calendar is: ", dt.now().isocalendar())
print("current iso format is: ", dt.now().isoformat())
print("current timestamp is: ", dt.now().timestamp())
```

```python
print("format date and time: ", dt.now().strftime("%Y-%m-%d %H:%M:%S"))
print("parse date and time: ", dt.strptime("2024-06-01 12:00:00", "%Y-%m-%d %H:%M:%S"))
print("get current date and time: ", dt.now())
print("current month name: ", dt.now().strftime("%B"))
print("current day name: ", dt.now().strftime("%A"))
print("current month name: ", dt.now().strftime("%b"))
print("current day name: ", dt.now().strftime("%a"))
print("current day of the year: ", dt.now().timetuple().tm_yday)
print("current week number: ", dt.now().isocalendar()[1])
```

```python
total_folders = 0
total_files = 0
for path,folders,files in os.walk(os.getcwd()):
    print("current path:",path)
    print("folders:",folders)
    print("files:",files)
    total_folders+=len(folders)
    total_files+=len(files)
print("total folders:",total_folders)
print("total files:",total_files)
```
