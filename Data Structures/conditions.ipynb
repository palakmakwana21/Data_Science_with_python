# Input 3 sides of traingle and check if it is 
- equilateral(all equal)
- isoceles(2 side equal ) 
- scalene(no side equal)

```python
a=int(input("enter no. a: "))
b=int(input("enter no. b: "))
c=int(input("enter no. c: "))
if a==b and b==c:
    print(f'{a} equilateral triangle')
elif a==b or b==c or a==c:
    print(f'{a} isosceles triangle')
else:
    print(f'{a} scalene triangle')
            
```

```python
# acuate, obstuse, right angled triangle
a=int(input("enter no. a: "))
b=int(input("enter no. b: "))
c=int(input("enter no. c: "))
if a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or c**2 + a**2 == b**2:
    print("right angled triangle")
elif a**2 + b**2 < c**2 or b**2 + c**2 < a**2 or c**2 + a**2 < b**2:
    print("obtuse angled triangle")
else:   
   print("acute angled triangle") 
   
```
