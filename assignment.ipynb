Question: Simple ATM Program
Create a Python program that simulates a simple ATM machine.
Requirements:
Start with an initial balance of ₹10,000.
Display the following menu repeatedly:
Check Balance
Deposit Money
Withdraw Money
Exit
If the user chooses:
Check Balance: Display the current balance.
Deposit: Ask for an amount and add it to the balance.
Withdraw: Ask for an amount and deduct it from the balance if sufficient funds are available.
Exit: Terminate the program.
If the user enters an invalid choice, display an appropriate message.
Use a while loop to keep showing the menu until the user chooses to exit.

```python
balance = 10000
while True:
 choice =input('enter a choice :\na -check balance, b - deposit,c- withdraw,d-exit\n')
 if choice =='a':
   print('your balance is :',balance)
 elif choice =='b':
   amt = int(input('enter the amount to deposit :'))
   balance+=amt
   print('your new balance is :',balance)
 elif choice =='c':
    amt = int(input('enter the amount to withdraw :'))
    balance -=amt
    print('your new balance is :',balance)
 elif choice =='d':
   print('thank u for using our service')
   break
 else:
   print("invalid choice,please try again")
   
```
