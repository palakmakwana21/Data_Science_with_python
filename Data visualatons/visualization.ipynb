```python
sns.palettes.SEABORN_PALETTES.keys()
```

```python
sns.countplot(titanic,x='embark_town', hue = 'who', palette ='muted')
```

```python
sns.palplot(sns.color_palette('muted'))
```

```python
palette_name = list(sns.palettes.SEABORN_PALETTES.keys())
print(palette_name)
for name in palette_name:
    sns.palplot(sns.color_palette(name))
# tell which color palette are there function palplot
```
