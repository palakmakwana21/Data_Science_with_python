```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = sns.load_dataset('iris')
df
```

```python
#histogram - distribution of sepal length
plt.hist(df['sepal_length'],color='skyblue', edgecolor='black')
plt.title('distribution of sepal length ')
plt.xlabel('sepal length')
plt.ylabel('frequency')
plt.show()
```

```python
#line plot - petal length of first 50 sample
plt.plot(df['petal_length'][:50], color='orange')
plt.title('petal length of first 50 samples')
plt.xlabel('sample index')
plt.ylabel('petal length')
plt.show()
```

```python
#calculate mean values for each species
mean_value = df.groupby('species').mean()
```

```python
mean_value.plot(kind ='bar', figsize=(5,5))
plt.title("average feature values per species")
plt.ylabel("average value")
plt.xticks(rotation=20)
plt.show()
```

```python
#creating a figure and axis object
fig, ax = plt.subplots(figsize=(5,3))
#setting the color of figure and axes
fig.set_facecolor('white')
ax.set_facecolor('lightgreen')
#plotting the data
x = [1,2,3,4,5,6,7,8,9]
y = [3,6,2,4,7,9,8,2,5]
ax.plot(x,y,
        color ='red',
        linewidth =1,
        linestyle='--',
        marker='o',
        markersize= 4,
        markerfacecolor='white',
        markeredgecolor ='black',
        markeredgewidth =2)

ax.set_title('customised line chart ', color ='red',fontsize= 18)
ax.set_xlabel("y- axis",color ='green',fontsize=15)
ax.set_ylabel("y- axis",color ='green',fontsize=15)

#add text
ax.text(3,7,'sample text')

ax.axhline(y=5, color ='black', alpha =0.2)
ax.axvline(x=5, color ='black', alpha =0.2)
 


```
